```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 45
plt.rcParams['font.family'] = 'Times New Roman'

# Read data from CSV
df = pd.read_csv('./val_acc.csv')
# Extract data for specified epoch ranges

# Extract data for specified epoch ranges
epoch_ranges = [(1, 15), (15, 30), (30, 50)]
dfs = [df[(df['Epoch'] >= start) & (df['Epoch'] <= end)] for start, end in epoch_ranges]

# Create a single subplot for all epoch ranges
fig, axes = plt.subplots(1, len(epoch_ranges), figsize=(40, 8))

# Plot for each epoch range
for i, (df_range, (start, end)) in enumerate(zip(dfs, epoch_ranges)):
    axes[i].plot(df_range['Epoch'], df_range['TrainAcc'], label='Train Accuracy', linewidth=5)
    axes[i].plot(df_range['Epoch'], df_range['TrainLoss'], label='Train Loss', linewidth=5)
    axes[i].plot(df_range['Epoch'], df_range['ValAcc'], label='Validation Accuracy', linewidth=5)
    axes[i].plot(df_range['Epoch'], df_range['ValLoss'], label='Validation Loss', linewidth=5)
    axes[i].set_title(f'Epochs {start}-{end}')
    axes[i].set_xlabel('Epoch')
    axes[i].set_ylabel('Value')
    # axes[i].legend()

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()

```
